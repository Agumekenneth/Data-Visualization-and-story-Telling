# Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data['x1'],data['y1'],color = 'blue')
m, c = np.polyfit(data['x1'], data['y1'], 1)   # slope and intercept
y_fit = m * data['x1'] + c

plt.plot(data['x1'], y_fit)
plt.title("Scatter plot of X1 against Y1")
plt.xlabel("x1")
plt.ylabel("y1")

plt.tight_layout()
plt.show()
```

- The scatter plot shows a positive relationship between X1 and Y1, where Y1 generally increases as X1 increases. The line of best fit shows a positive linear trend.

```python
plt.figure(figsize=(8,6))
plt.scatter(data['x2'],data['y2'],color = 'blue')
m, c = np.polyfit(data['x2'], data['y2'], 1)   # slope and intercept
y_fit = m * data['x1'] + c

plt.plot(data['x2'], y_fit)
plt.title("Scatter plot of X2 against Y2")
plt.xlabel("x2")
plt.ylabel("y2")

plt.tight_layout()
plt.show()
```

- The scatter plot shows a positive linear relationship between x and y. As X increases, Y also generally increases and line of best fit slopes upward showing a positive correlation. There seems to be a point farthest from the line of best fit at x=4; other than that most of the points are just scattered around the line suggesting a moderately strong relationship between the variables
