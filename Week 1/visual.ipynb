```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data['x1'],data['y1'],color = 'blue')
m, c = np.polyfit(data['x1'], data['y1'], 1)   # slope and intercept
y_fit = m * data['x1'] + c

plt.plot(data['x1'], y_fit)
plt.title("Scatter plot of X1 against Y1")
plt.xlabel("x1")
plt.ylabel("y1")

plt.tight_layout()
plt.show()
```
