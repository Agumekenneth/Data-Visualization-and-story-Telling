```python
import pandas as pd
import matplotlib.pyplot as plt
divers = pd.read_csv('deepest-diving-animals.csv')
divers #display the dataframe
```

```python
divers['category'].value_counts() #count the number of occurrences of each category
```

```python
max_depth = divers.groupby('category')['depth'].max() #group the dataframe by category and find the maximum depth for each category
max_depth
```

```python
df = max_depth.reset_index(name='max_depth') #reset the index of the dataframe and rename the column to max_depth
df
```

```python
plt.figure(figsize=(6,3))
plt.barh(df['category'], df['max_depth'])
plt.ylabel('Maximum depth (m)')
plt.show()
```

```python
df = df.sort_values('max_depth')
df
```

```python
plt.figure(figsize=(6,3))
plt.barh(df['category'],df['max_depth'])
plt.ylabel('Maximum depth (m)')

ax =plt.gca() #get the current axes
ax.spines[['top','right','bottom','left']].set_visible(False) #hide the top, right and bottom spines
ax.grid(axis='x', color='black', alpha=0.5)
ax.tick_params(axis='both',length=0) #remove the tick marks
plt.show()
```

```python
def clean ():
    ax =plt.gca()
    ax.spines[['top','right','bottom','left']].set_visible(False)
    ax.grid(axis='x',color='black',alpha=0.5)
    ax.tick_params(axis='both',length=0) #remove the tick marks
```

```python
plt.figure(figsize=(6,3))
plt.barh(df['category'], df['max_depth'])
plt.ylabel('Maximum depth (m)')
clean()
plt.show()
```
